```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('nba_playoff_series.csv')

# Rename columns for better readability
# Adjust these names if your CSV headers differ
df.columns = ['Year', 'League', 'Series', 'Dates', 'Empty1', 'Team1', 'W1', 'Empty2', 'Team2', 'W2', 'Empty3', 'Favorite', 'Underdog']

# Function to compute total games for teams reaching finals
def calculate_games(year, winner, loser):
    games_winner = 0
    games_loser = 0
    
    # Filter all matches for the given year
    year_matches = df[df['Year'] == year]
    
    # Iterate over each round
    for _, row in year_matches.iterrows():
        if winner in row['Team1'] or winner in row['Team2']:
            games_winner += row['W1'] + row['W2']
        if loser in row['Team1'] or loser in row['Team2']:
            games_loser += row['W1'] + row['W2']
    
    # Exclude the finals from total games
    finals = year_matches[year_matches['Series'] == 'Finals']
    finals_row = finals.iloc[0]
    
    if winner in finals_row['Team1'] or winner in finals_row['Team2']:
        games_winner -= finals_row['W1'] + finals_row['W2']
    if loser in finals_row['Team1'] or loser in finals_row['Team2']:
        games_loser -= finals_row['W1'] + finals_row['W2']
    
    return f"{year}, {games_winner}, {games_loser}, {games_winner - games_loser}"

# Extract all final matches
finals_matches = df[df['Series'] == 'Finals']

# Initialize list to store all results
all_results = []

# Iterate over each final match to calculate total games
for _, row in finals_matches.iterrows():
    year = row['Year']
    winner = row['Team1']
    loser = row['Team2']
    result = calculate_games(year, winner, loser)
    all_results.append(result)

# Print all results
for res in all_results:
    print(res)
```
